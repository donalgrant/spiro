```python
f=SpiroFig()
sd=SpiroData()
np.random.seed(11)
ring=10
a=2.5
bv=[0.9*a+0.03*a*i for i in range(10)]
x0=np.random.standard_normal(40)*40-20
y0=np.random.standard_normal(40)*40-20
for o in range(x0.shape[0]):
    o0=np.random.uniform(0,2*pi)
    of0=np.random.uniform(2*pi)
    for i in range(np.random.randint(3,10)): 
        l=max(0.01,0.26+0.15*np.random.standard_normal(1)[0])
#        print(o,i,l)
        sd.add(spiro_arc(x0=x0[o],y0=y0[o],orient=o0,
                         R=ring,wheel=Wheel(a,bv[i],of0+pi/5*i),loops=l))
f.plot(sd,color_scheme='time',cmap='hot')
```
