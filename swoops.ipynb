```python
from swoops import swoops
f = SpiroFig()
path='out/Swoops/'
fig_number=0
```

```python
f.plot(swoops(),color_scheme='time',cmap='hot')
f.save_fig(path+f'Swoop-{fig_number}.png')
fig_number += 1
```

```python
f.plot(swoops(wheel_e=0.8,yscale=20,xscale=20,
              swoop_min=0.03,swoop_mean=0.1,swoop_spread=0.1),
       color_scheme='x-y',cmap='inferno')
f.save_fig(path+f'Swoop-{fig_number}.png')
fig_number += 1
```

```python
f.plot(swoops(seed=11,ring_e=0.8,xscale=20,yscale=15,swoop_min=0.03,swoop_mean=0.1,swoop_spread=0.2),color_scheme='x-y',cmap='inferno')
f.save_fig(path+f'Swoop-{fig_number}.png')
fig_number += 1
```

```python
sd = SpiroData()

sd.add(swoops(n=40,ring_e=0.8,exponential=True,xscale=50,yscale=10,swoop_min=0.03,swoop_mean=0.1,swoop_spread=0.1))
f.plot(sd,color_scheme='x-y',cmap='inferno')
f.plot(sd,color_scheme='x-y',cmap='inferno',dot_size=100,alpha=0.03,new_fig=False,subsample=100)
sd.reset()
sd.add(swoops(n=15,wheel_e=0.8,yscale=20,xscale=20,
              swoop_min=0.03,swoop_mean=0.1,swoop_spread=0.1))
f.plot(sd,color_scheme='x-y',cmap='inferno',alpha=0.3,new_fig=False)
sd.reset()
sd.add(swoops(n=10,yscale=30))
f.plot(sd,color_scheme='x-y',cmap='inferno',alpha=0.1,new_fig=False)
f.save_fig(path+f'Swoop-{fig_number}.png')
fig_number += 1
```
