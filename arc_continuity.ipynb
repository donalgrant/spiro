This notebook is being used to work out the continuity across an arc.  May also use it to refactor "spiro" into a shortcut for spiro_arc.

```python
from spiro import *
```

```python
F = SpiroFig()
S = SpiroData()
```

```python
S.reset()
F.plot(spiro_arc(orient=0,R=10,a=0.01,b=0,loops=1,offset=0,invert=False,reverse=False),cmap='hsv',color_scheme='time',new_fig=True)
o=0
r=10
for i in range(1):
    S.add(spiro_arc(orient=0,R=r,a=3.5,b=3.3,loops=1,offset=o,invert=False,reverse=False))
    S.add(rotate(0,r,S.xc(),S.yc(),pi))
    S.add(spiro_arc(orient=0,R=10,a=3.5,b=3.3,loops=1,offset=S.pc(),invert=True,reverse=True))
    S.add(rotate(0,r,S.xc(),S.yc(),pi))
    o=S.pc()
    
F.plot(S,new_fig=False,cmap='autumn',color_scheme='time')
#F.plot(spiro_arc(orient=pi,R=10,a=1,b=2,loops=.25,offset=0,invert=True,reverse=True),new_fig=False,cmap="autumn",color_scheme='radial')
```
