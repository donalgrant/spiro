This notebook is being used to work out the continuity across an arc.  May also use it to refactor "spiro" into a shortcut for spiro_arc.

```python
from spiro import *
```

```python
F = SpiroFig()
S = SpiroData()
```

```python
S.reset()
r=10
a=1.3
p=2.3
F.plot(spiro_arc(orient=0,R=r,a=0.01,b=0,loops=1,offset=0,invert=False,reverse=False),cmap='hsv',color_scheme='time',new_fig=True)
o=0
for i in range(1):
    S.add(spiro_arc(orient=0,R=r,a=a,b=p,loops=0.25,offset=o+pi,invert=True,reverse=True))
    o=S.pc()
    S.add(rotate(-r,0,S.xc(),S.yc(),pi))
    S.add(spiro_arc(orient=-pi/2,R=r,a=a,b=p,loops=0.25,offset=o+pi,invert=False,reverse=False))
    o=S.pc()
    S.add(rotate(0,r,S.xc(),S.yc(),pi))
    
F.plot(S,new_fig=False,cmap='autumn',color_scheme='time')
```

```python
S.reset()
r=9
F.plot(spiro_arc(orient=-pi/2,R=r,a=0.01,b=0,loops=1,offset=0,invert=False,reverse=False),
       cmap='hsv',color_scheme='time',new_fig=True)
o=0
rot=pi/10
(a,b)=(1.5,1.3)
for i in range(1):
    S.add(spiro_arc(orient=-pi/2+rot,R=r,a=a,b=b,loops=0.5,offset=o,invert=False,reverse=False))
    S.add(spiro_arc(orient=-pi/2+rot,R=r,a=a,b=b,loops=0.5,offset=o,invert=False,reverse=True))
    S.add(spiro_arc(orient=-pi/2+rot,R=r,a=a,b=b,loops=0.5,offset=o,invert=True,reverse=False))
    S.add(spiro_arc(orient=-pi/2+rot,R=r,a=a,b=b,loops=0.5,offset=o,invert=True,reverse=True))
    
F.plot(S,new_fig=False,cmap='autumn',color_scheme='time')
```

```python
(x1,y1,x2,y2)=(0,0,30,0)
(a,b)=(2,2)
S.reset()
S.add(roll(x1,y1,x2,y2,0.01,0.0,offset=0,invert=False))
S.add(roll(x1,y1,x2,y2,a,b,offset=0,invert=False))
o=S.pc()
S.add(rotate(x2,y2,S.xc(),S.yc(),pi))
S.add(roll(x2,y2,x1,y1,a,b,offset=o+pi,invert=False))
S.add(rotate(x1,y1,S.xc(),S.yc(),pi))
S.add(roll(x1,y1,x2,y2,a,b,offset=0,invert=True))
o = S.pc()
S.add(rotate(x2,y2,S.xc(),S.yc(),-pi))
S.add(roll(x2,y2,x1,y1,a,b,offset=o-pi,invert=True))
S.add(rotate(x1,y1,S.xc(),S.yc(),-pi))
F.plot(S)
```
