```python
sd = SpiroData()
offset=0
c="inferno"
cs="radial"
w=2.2
pen=1
pen_inc=0.5
depth=-3
for j in range(5):
    for i in range(10):
        sd.add(roll(20,depth,0,20,w,pen+j*pen_inc,offset=offset))
        sd.add(rotate(0,20,sd.xc(),sd.yc(),pi))
        sd.add(spiro_arc(10,20,-pi/2,10,w,pen+j*pen_inc,loops=0.5,offset=sd.pc()))
        sd.add(rotate(20,20,sd.xc(),sd.yc(),pi/2))
        sd.add(spiro_arc(30,20,-pi/2,10,w,pen+j*pen_inc,loops=0.5,offset=sd.pc()))
        sd.add(rotate(40,20,sd.xc(),sd.yc(),pi))
        sd.add(roll(40,20,20,depth,w,pen+j*pen_inc,offset=sd.pc()))
        sd.add(rotate(20,depth,sd.xc(),sd.yc(),pi/2))
        offset=sd.pc()
        
offset=0
c="inferno"
cs="radial"
w=1.5
pen=w*1.1
pen_inc=0.5
depth=-3
for j in range(1):
    for i in range(22):
        sd.add(roll(20,depth,0,20,w,pen+j*pen_inc,offset=offset,invert=True))
        sd.add(spiro_arc(10,20,-pi/2,10,w,pen+j*pen_inc,loops=0.5,offset=sd.pc(),invert=True))
        sd.add(spiro_arc(30,20,-pi/2,10,w,pen+j*pen_inc,loops=0.5,offset=sd.pc(),invert=True))
        sd.add(roll(40,20,20,depth,w,pen+j*pen_inc,offset=sd.pc(),invert=True))
        offset=sd.pc()
    
f.plot(sd,cmap=c,color_scheme=cs)
```
