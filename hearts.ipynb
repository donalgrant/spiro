```python
f = SpiroFig()
f.add(spiro_steps(10.0,-6.0,9.0,1.4,10,offset=pi/40))
f.plot()
```

```python
f.reset_data()
f.clear_fig()
offset=0
c="inferno"
cs="radial"
w=2.2
pen=1
pen_inc=0.5
depth=-3
for j in range(5):
    for i in range(10):
        f.add(roll(20,depth,0,20,w,pen+j*pen_inc,offset=offset))
        f.add(rotate(0,20,f.xc(),f.yc(),pi))
        f.add(spiro_arc(10,20,-pi/2,10,w,pen+j*pen_inc,loops=0.5,offset=f.pc()))
        f.add(rotate(20,20,f.xc(),f.yc(),pi/2))
        f.add(spiro_arc(30,20,-pi/2,10,w,pen+j*pen_inc,loops=0.5,offset=f.pc()))
        f.add(rotate(40,20,f.xc(),f.yc(),pi))
        f.add(roll(40,20,20,depth,w,pen+j*pen_inc,offset=f.pc()))
        f.add(rotate(20,depth,f.xc(),f.yc(),pi/2))
        offset=f.pc()
        
offset=0
c="inferno"
cs="radial"
w=1.5
pen=w*1.1
pen_inc=0.5
depth=-3
for j in range(1):
    for i in range(22):
        f.add(roll(20,depth,0,20,w,pen+j*pen_inc,offset=offset,invert=True))
        f.add(spiro_arc(10,20,-pi/2,10,w,pen+j*pen_inc,loops=0.5,offset=f.pc(),invert=True))
        f.add(spiro_arc(30,20,-pi/2,10,w,pen+j*pen_inc,loops=0.5,offset=f.pc(),invert=True))
        f.add(roll(40,20,20,depth,w,pen+j*pen_inc,offset=f.pc(),invert=True))
        offset=f.pc()
    
f.plot(cmap=c,color_scheme=cs)
```
