```python
f=SpiroFig()
sd=SpiroData()
np.random.seed(12)
ring=10
a=2.5
e=0.8
n=40
bv=[0.9*a+0.03*a*i for i in range(10)]
x0=np.random.standard_normal(n)*2*ring-ring
y0=np.random.standard_normal(n)*2*ring-ring
for o in range(x0.shape[0]):
    o0=np.random.uniform(0,2*pi)
    of0=np.random.uniform(2*pi)
    for i in range(np.random.randint(3,10)): 
        l=max(0.01,0.2+0.05*np.random.standard_normal(1)[0])
#        print(o,i,l)
        sd.add(elliptical_arc(x0=x0[o],y0=y0[o],orient=o0,
                         R=ring,wheel=Ellipse(a,e,bv[i],of0+pi/5*i),loops=l))
f.plot(sd,color_scheme='x-y',cmap='inferno',dot_size=0.1)
```

```python
sd.reset()
for p in range(10):
    sd.add(elliptical_arc(x0=0,y0=0,orient=0,R=2.0,wheel=Ellipse(9.5,0.4,5+p/2,0),
                        loops=10,spacing=pi/4000,inside=False))
f.plot(sd,cmap='tab20b',color_scheme='time',alpha=1.0)
f.caption('outside R=2, e=9.5/0.4, step pen from 5 to 10')
f.save_fig('Ellipse-A.png')
```

```python
sd.reset()
for j in range(10):
    sd.add(elliptical_arc(x0=-j/5,y0=j/5,orient=0,R=2.0,wheel=Ellipse(6.5,0.8,2,pi/20*j),
                        loops=3,spacing=pi/4000,inside=False))
f.plot(sd,cmap='Dark2',color_scheme='time',dot_size=.1,alpha=1.0)
f.caption('outside R=2.0 (origin stepped), e=6.5/0.3/2 offset stepped')
f.save_fig('Ellipse-B.png')
```
